```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('yoshi.png')
h,w,_ = img.shape
img.shape
```

```python
alpha = np.pi / 8
zoom = 6
zx, zy = np.cos(alpha)*zoom, np.sin(alpha)*zoom
zx, zy
```

```python
hcut = np.arange(0, h)
wcut = np.arange(0, w)
xv, yv = np.meshgrid(wcut, hcut)
r = 10

xv = r*(xv - w/2)/w
yv = r*(yv - h/2)/h

# xt = xv * zx - yv * zy
# yt = xv * zy + yv * zx
xt = xv * xv - yv * yv
yt = xv * yv + yv * xv

xv = (xt*w/r + w/2).astype(int) % w
yv = (yt*h/r + h/2).astype(int) % h
```

```python
fig = plt.figure()
fig.set_size_inches(24, 12)

fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(img)

fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(img[yv, xv])

# imgplot = plt.imshow(img[:,:,0], cmap="hot")
# imgplot.set_clim(0.0, 1.0)
# plt.colorbar(orientation='horizontal')
```
